```python
import pandas as pd
```

```python
stats = pd.read_csv("data/2019-La-Liga.csv")
```

```python
#Find the total number of goals scored by each team (Performance|Gls)
stats.groupby("Team")["Performance|Gls"].sum()
```

```python
players = ["Loren Morón","Ángel Rodríguez","Luis Suárez", "Karim Benzema", "Lionel Messi"]
subset = stats[stats["Player"].isin(players)]
subset.groupby(["Player"])[["Performance|Gls","Performance|Ast"]].sum().plot(kind="barh", stacked=True)
```

```python
df = stats.groupby("Player")[["Performance|Gls", "Min"]].sum()
df = df[df["Performance|Gls"] >= 10]
df["Minutes_Per_Goal"] = df["Min"] / df["Performance|Gls"]
# df.apply(lambda x: [1, 2], axis=1, result_type='expand')
df.sort_values("Minutes_Per_Goal")
```

```python
stats["Age_yrs"] =  stats["Age"].str.split("-").str[0].astype("float")
```

```python
stats["Age_days"] =  stats["Age"].str.split("-").str[1].astype("float")
```

```python
stats["Age_in_day"] = stats["Age_yrs"] * 365 + stats["Age_days"]
```

```python
stats[["Age", "Age_yrs", "Age_days", "Age_in_day"]].sort_values("Age_in_day")
```

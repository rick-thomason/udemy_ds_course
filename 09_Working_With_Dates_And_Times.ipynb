```python
ufos[ufos['date_time'] < '1980']['date_time'].dt.year.value_counts().sort_index()
```

```python
ufos['date_time'].dt.date.value_counts()
```

```python
ufos['posted'] = pd.to_datetime(ufos['posted'])
```

```python
ufos.info()
```

```python
ufos['how_long_to_report'] = ufos['posted'] - ufos['date_time']
```

```python
ufos.head()
```

```python
ufos.rename(columns={'how_long_to_report': 'time_to_report'}, inplace=True)
```

```python
ufos.info()
```

```python
longest_times_to_report = ufos.sort_values(['time_to_report'], ascending=False).head(10)
```

```python
longest_times_to_report
```

```python
round(longest_times_to_report['time_to_report'].dt.days / 365, 2)
```

```python
ufos['days_between_reporting'] = ufos['time_to_report'].dt.days
```

```python
ufos.drop(columns='time_to_report')
ufos.head()
```

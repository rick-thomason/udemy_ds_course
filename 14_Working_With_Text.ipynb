## Replace( ) 

```python
ufos = pd.read_csv('data/nuforc_reports.csv')
ufos.head()
```

```python
ufos['duration'].str.replace('seconds', 's')
```

```python
def abbrv(regular_expression_object):
    return regular_expression_object.group()[0]
```

```python
ufos['duration'].str.replace('seconds|minutes', abbrv, regex=True)
```

## Contains( )

```python
ufos['duration'].str.contains('hour').value_counts()
```

```python
ufos[ufos['duration'].str.contains('hour', na=False)]
```

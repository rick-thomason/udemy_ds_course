# Bar Plots

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6,5), dpi=100)
plt.bar(plants, germinated, color='orchid', label='Germinated', width=0.6)
plt.title('Germination Rates')
plt.ylabel('Number of Seedlings', labelpad=20)
plt.show()
```

```python
plt.figure(figsize=(6,5), dpi=100)
plt.bar(plants, died, color='red', label='Died', bottom=germinated)
plt.bar(plants, germinated, color='blue', label='Germinated')
plt.title('Germination Rates')
plt.xlabel('Plants', labelpad=20)
plt.legend(shadow=True, facecolor='white')
plt.show()
```

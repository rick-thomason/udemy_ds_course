# Adding & Removing Columns

```python
import pandas as pd
houses = pd.read_csv("data/kc_house_data.csv")
titanic = pd.read_csv("data/titanic.csv")
netflix = pd.read_csv("data/netflix_titles.csv", sep="|", index_col=0)
btc = pd.read_csv("data/coin_Bitcoin.csv")
countries = pd.read_csv("data/world-happiness-report-2021.csv")
```

```python
countries.columns = [name.lower() for name in countries.columns]
btc.columns = [name.lower() for name in btc.columns]
```

## Dropping Rows/Cols

```python
btc.head()
```

```python
# drop name and symbol columns
btc = btc.drop(labels=['name', 'symbol', 'sno'], axis='columns')
```

```python
btc.head()
```

```python
# dropping a row from a dataframe
countries.drop(labels='Denmark')
```

```python
# drop rows by position
countries.drop(countries.index[0:3]) # drops first three in df
```
